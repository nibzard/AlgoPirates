```
# Run me!  See today's paper sheet for the plan.
import turtle
t = turtle.Turtle()
size = 40
colors = ['red','white','blue']
for row in range(3):
    t.up(); t.goto(-60, 60 - row*size); t.down()
    t.fillcolor(colors[row]); t.begin_fill()
    for _ in range(2):
        t.forward(120); t.right(90); t.forward(size); t.right(90)
    t.end_fill()
### TODO ###
# Change colors or sizes to make your own flag!
t.hideturtle()
turtle.done()
```
